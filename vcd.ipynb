```python
from pandas import DataFrame
import matplotlib.pyplot as plt

spectrum_curve = 'spectrum_curve_vcd.txt'
spectrum_line = 'spectrum_line_vcd.txt'

sc = DataFrame(columns=['x','y'])
sl = DataFrame(columns=['x','y'])
```

```python
arq = open(spectrum_curve, 'r').readlines()

for i in range(len(arq)):
    x = float(arq[i].split()[0])
    y = float(arq[i].split()[1])

    sc = sc._append({'x': x, 'y': y}, ignore_index=True)
```

```python
arq2 = open(spectrum_line, 'r').readlines()

for j in range(len(arq2)-1):
    x = float(arq2[j].split()[0])
    y = float(arq2[j].split()[1])

    sl = sl._append({'x': x, 'y': y}, ignore_index=True)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)

ax.plot(sc['x'].values, sc['y'].values, color='#0005FF')
ax.fill_between(sc['x'].values, sc['y'].values, color='#0005FF', alpha=0.3)
ax.set_ylabel(r'$\mathbf{\Delta \epsilon}$' +' (arb.)', fontsize=14, fontweight='bold')
ax.set_xlabel('Número de Onda ($\mathbf{cm^{-1}}$)', fontsize=14, fontweight='bold')
ax.set_xlim(min(sc['x']), max(sc['x']))

ax.hlines(0, min(sc['x']), max(sc['x']), color='black', linestyle='--', alpha=0.5)

ax.set_xlim(0,3500)
    
ax.invert_xaxis()
ax.grid(linestyle='--', color='black', alpha=0.3)
plt.savefig('VCD.png', dpi=300)
```
