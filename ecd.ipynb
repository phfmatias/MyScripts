```python
from pandas import DataFrame
import matplotlib.pyplot as plt

spectrum_curve = 'spectrum_curve_ecd.txt'
spectrum_line = 'spectrum_line_ecd.txt'

sc = DataFrame(columns=['x','y'])
sl = DataFrame(columns=['x','y'])
```

```python
arq = open(spectrum_curve, 'r').readlines()

for i in range(len(arq)):
    x = float(arq[i].split()[0])
    y = float(arq[i].split()[1])

    sc = sc._append({'x': x, 'y': y}, ignore_index=True)
```

```python
arq2 = open(spectrum_line, 'r').readlines()

for j in range(len(arq2)-1):
    x = float(arq2[j].split()[0])
    y = float(arq2[j].split()[1])

    sl = sl._append({'x': x, 'y': y}, ignore_index=True)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)

# ax2 = ax.twinx()
# ax2.plot(sl['x'].values, sl['y'].values, color='#000000', zorder=1)
# ax2.set_ylim(ax.get_ylim())
# ax2.set_ylabel('Força de Rotação (cgs)', fontsize=14, fontweight='bold')

ax.plot(sc['x'].values, sc['y'].values, color='#0005FF')
ax.fill_between(sc['x'].values, sc['y'].values, color='#0005FF', alpha=0.3)
ax.set_ylabel(r'$\mathbf{\Delta \epsilon}$' +' (arb.)', fontsize=14, fontweight='bold')
ax.set_xlim(min(sc['x']), max(sc['x']))
ax.set_xlabel('Comprimento de Onda (nm)', fontsize=14, fontweight='bold')

ax.hlines(0, min(sc['x'].values), max(sc['x'].values), color='black')

# if abs(min(sc['y'])) > abs(max(sc['y'])):
#     ax.set_ylim(-abs(min(sc['y']))-100, abs(min(sc['y']))+100)
# else:
#     ax.set_ylim(-abs(max(sc['y']))-100, abs(max(sc['y']))+100)

# if abs(min(sl['y'])) > abs(max(sl['y'])):
#     ax2.set_ylim(-abs(min(sl['y']))-1000, abs(min(sl['y']))+1000)
# else:
#     ax2.set_ylim(-abs(max(sl['y']))-1000, abs(max(sl['y']))+1000)

ax.grid(linestyle='--', color='black', alpha=0.3)
plt.savefig('ECD.png', dpi=300)
```
